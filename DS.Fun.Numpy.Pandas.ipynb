```python
### script for checking age eligibility for vote 
def Vote(n):
    if n>=18:
        return "You can Vote"
    else:
        return " Your Not Eligible for Vote"
n=int(input("enter your age:"))
print(Vote(n))
```

```python
### script for Grade 
def Grade(marks):
    if marks>=90:
        return "A"
    elif 75<marks<89:
        return "B" 
    elif 50<marks<74:
        return "C" 
    else:
        return "Fail" 
marks=int(input("enter marks:"))
print(Grade(marks))
```

```python
### scrip for positive or negative number 
def Sign(n):
    if n>0: 
        return 'positive' 
    elif n<0: 
        return 'negative' 
    else:
        return 'zero' 
n=int(input("enter number:"))
print(Sign(n))
```

```python
### creating a small dataframe with rows and 4 columns
### create a dictionary
data = {
    "Name": ["Rakesh", "Anita", "Vijay", "Sneha", "Arjun", 
             "Rakesh", "Meera", "Kiran", "Priya", "Rahul"],
    "Age": [25, 30, 28, 22, 35, 25, 29, 31, 26, 24],
    "City": ["Bengaluru", "Delhi", "Mumbai", "Chennai", "Hyderabad", 
             "Bengaluru", "Pune", "Kolkata", "Jaipur", "Goa"],
    "Salary": [50000, 60000, np.nan, 45000, 70000, 50000, 55000, 62000, np.nan, 48000]
}
```

```python
#### convert dict into dataframe 
df=pd.DataFrame(data)
```

```python
df
```

```python
### info() 
df.info()
```

```python
#### describe() 
df.describe()
```

```python
### selecting name and city columns
df[['Name','City']]
```

```python
### drop the city column 
df=df.drop(columns=['City'],axis=1)
```

```python
df
```

```python
### missing values 
df.isnull().sum()
```

```python
### filling the missing values respected value
df.fillna({'Salary':df['Salary'].median()},inplace=True)
```

```python
df.isnull().sum()
```

```python
#### duplicates finding 
df.duplicated().sum()
```

```python
#### remove duplicates rows 
df.drop_duplicates(inplace=True,ignore_index=True)
```
